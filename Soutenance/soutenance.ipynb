## dataframe experience junior 

```python
filterEN = df[(df.experience_level == "EN")]
```

### describe dataframe experience junior

```python
filterEN.describe()
```

### histogramme dataframe experience junior - company location

```python
sns.histplot(data=filterEN, x="company_location")
```

### histogramme dataframe experience junior - salary in usd

```python
sns.histplot(data=filterEN, x="salary_in_usd") 
```

### histogramme dataframe experience junior - job title



```python
sns.histplot(data=filterEN, y="job_title")
```

### catplot dataframe experience junior - job title par année



```python
filterEN = df[(df.experience_level == "EN")]
sns.catplot(y="job_title", x="work_year", data=filterEN, height= 10)
```

### relplot dataframe experience junior - salaire selon job title en 2021



```python
filterEN = df[(df.experience_level == "EN") & (df.work_year > 2020) & (df.work_year < 2022)]
sns.relplot(
    data=filterEN,
    y="job_title", x="salary_in_usd", height= 10
)
```

### relplot dataframe experience junior - salaire selon job title en 2022



```python
filterEN = df[(df.experience_level == "EN") & (df.work_year > 2021)]
sns.relplot(
    data=filterEN,
    y="job_title", x="salary_in_usd", height= 10
)
```

### barplot dataframe experience junior - salaire moyen selon job title



```python
filterEN = df[(df.experience_level == "EN")]
sns.barplot(
    data=filterEN,
    y="job_title", x="salary_in_usd"
)
```
