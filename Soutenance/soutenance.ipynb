```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib as mpot
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
csvFilePath = "salaries_cyber.csv"

df = pd.read_csv(csvFilePath)
df.describe()
```

```python
filterEN = df[(df.experience_level == "EN")]
filterEN.describe()
```

```python
filterEN.describe()
```

```python
filterEN = df[(df.experience_level == "EN")]
sns.histplot(data=filterEN, x="salary_in_usd") 
```

```python
filterEN = df[(df.experience_level == "EN")]
sns.histplot(data=filterEN, y="job_title")
```

```python
filterEN = df[(df.experience_level == "EN") & (df.work_year > 2020) & (df.work_year < 2022)]
sns.relplot(
    data=filterEN,
    y="job_title", x="salary_in_usd", height= 10
)
```

```python
filterEN = df[(df.experience_level == "EN") & (df.work_year > 2021)]
sns.relplot(
    data=filterEN,
    y="job_title", x="salary_in_usd", height= 10
)
```

```python
filterEN = df[(df.experience_level == "EN") & (df.work_year < 2021)]
sns.relplot(
    data=filterEN,
    y="job_title", x="salary_in_usd", height= 10
)
```

```python
filterEN = df[(df.experience_level == "EN")]
sns.barplot(
    data=filterEN,
    y="job_title", x="salary_in_usd"
)
```

```python
filterEN = df[(df.experience_level == "EN")]
sns.catplot(y="job_title", x="work_year", data=filterEN, height= 15)
```
